```python
def check_input(element):
    return isinstance(element, str)

def get_substring(s):
    if not s:
        return ""

    char_index = {}
    start = 0
    max_len = 0
    max_substr = ""

    for i, char in enumerate(s):
        if char in char_index and char_index[char] >= start:
            start = char_index[char] + 1

        char_index[char] = i

        current_len = i - start + 1
        if current_len > max_len:
            max_len = current_len
            max_substr = s[start:i+1]

    return max_substr

def main(lst):
    strings_with_indices = [(i, s) for i, s in enumerate(lst) if check_input(s)]

    if not strings_with_indices:
        print(-1, '')
        return

    longest_index, longest_string = max(strings_with_indices, key=lambda x: len(x[1]))
    substring = get_substring(longest_string)

    print(longest_index, substring)
```
